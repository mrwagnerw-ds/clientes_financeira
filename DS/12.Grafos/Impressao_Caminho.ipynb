```python
from igraph import Graph, plot
```

```python
# Criação do grafo direcionado com peso entre as arestas.

grafo = Graph(edges = [(0,2), (0,1), (1,4), (1,5), (2,3), (6,7),(3,7), (4,7), (5,6)], directed = True)
grafo.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
grafo.es['weight'] = [1,2,3,4,5,3,2,1]
```

```python
# Visualização.

plot(grafo, bbox = (0,0,600,600), edge_label = grafo.es['weight'])
```

```python
# O caminho entre os vértices A - H.

caminho_vertice = grafo.get_shortest_paths(0,7, output = 'vpath')
caminho_vertice
```

```python
# Retorna a aresta que liga dois vértices.

caminho_aresta = grafo.get_shortest_paths(0,7, output = 'epath')
caminho_aresta
```

```python
# Mostra o id dos vértices que fazem parte do caminho.

aresta_id = []
for a in caminho_aresta[0]:
    aresta_id.append(a)
aresta_id
```

```python
# Nome dos vértices que fazem parte do caminho.

nome_vertice = []
for n in caminho_vertice[0]:
    print(grafo.vs[n]['label'])
    nome_vertice.append(grafo.vs[n]['label'])

nome_vertice
```

```python
# Colorir os vértices do caminho.

for v in grafo.vs:
    if v['label'] in nome_vertice:
        v['color'] = 'green'
    else:
        v['color'] = 'gray'
```

```python
# Colorindo as arestas do caminho.

for e in grafo.es:
    if e.index in aresta_id:
        e['color'] = 'blue'
    else:
        e['color'] = 'gray'
    
```

```python
plot(grafo, bbox = (0,0,500,500))
```
