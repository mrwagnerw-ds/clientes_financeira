```python
from igraph import Graph, plot
```

```python
# Criação dos grafos com pesos nas relações.

grafo5 = Graph(edges = [(0,1), (2,3), (0,2), (0,3)], directed = True)
grafo5.vs['label'] = ['Fernando', 'Pedro', 'José', 'Antonio']
grafo5.vs['peso'] = [20,30,40,50]
grafo5.es['TipodeAmizade'] = ['Amigo', 'Inimigo', 'Inimigo', 'Amigo']
grafo5.es['weight'] = [1,2,1,4]
print(grafo5)
```

```python
# Visualização das informações sobre os vértices.

for i in grafo5.vs:
    print(i)
```

```python
# Visualização das informações das arestas.

for i in grafo5.es:
    print(i) 
```

```python
# Definindo cores para as vértices.

grafo5.vs['cor'] = ['blue', 'red', 'green', 'gray']
plot(grafo5, bbox = (0,0,400,400), vertex_color = grafo5.vs['cor'])
```

```python
# Peso para as arestas.

plot(grafo5, bbox = (0,0,400,400), edge_width = grafo5.es['weight'], vertex_color = grafo5.vs['cor'])
```

```python
# Pesos para os vértices.

plot(grafo5, bbox = (0,0,400,400), vertex_size = grafo5.vs['peso'], edge_width = grafo5.es['weight'], vertex_color = grafo5.vs['cor'],
     margin = 50)
```

```python
# Curvatura.

plot(grafo5, bbox = (0,0,600,600), edge_width = grafo5.es['weight'], vertex_size = grafo5.vs['peso'],
     vertex_color = grafo5.vs['cor'], edge_curved = 0.8, margin =80)
```

```python
# Formato.

plot(grafo5, bbox = (0,0,600,600), vertex_color = grafo5.vs['cor'], vertex_size = grafo5.vs['peso'],
     edge_width = grafo5.es['weight'], edge_curved = 0.8, vertex_shape = ['square'], margin = 50)
```
