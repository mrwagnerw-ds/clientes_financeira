```python
import igraph
from igraph import Graph, plot
```

```python
# Carregamento de grafo no formato graphml.

grafo = igraph.load('Grafo.graphml')
print(grafo)
```

```python
plot(grafo, bbox = (0,0,600,600), margin = 90)
```

```python
# Visualizando o grau de entrada + saída.

grafo.degree(mode = 'all')
```

```python
# Visualizando grau de entrada e saída, separadamente.

grafo.degree(mode = 'in')
grafo.degree(mode = 'out')
```

```python
# Obtendo e imprimindo somente os graus de entrada.

grau = grafo.degree(mode = 'in')
print(grau)
```

```python
# Gerando um grafo com o vértice proporcional ao grau.

plot(grafo, vertex_size = grau, bbox = (0,0,800,800), margin = 120)
```

```python
# Retorna o diâmetro entre os grafos (maior distância entre os vértices).

grafo.diameter(directed = True)
```

```python
# Retorno dos vértices que possuem a maior distância entre os pontos do grafo.

grafo.get_diameter()
```

```python
# Retorno dos vizinhos de cada vértice.

grafo.neighborhood()
```

```python
# Verificar se o grafo é isomórfico.

grafo2 = grafo
grafo.isomorphic(grafo2)
```
