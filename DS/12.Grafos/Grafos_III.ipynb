```python
from igraph import Graph, plot
```

```python
# Recriando o grafo 4.

grafo4 = Graph(directed = False)
grafo4.add_vertices(5)
grafo4.add_edges([(0,1), (1,2), (2,3), (3,4), (4,0), (0,2), (2,1)])
grafo4.add_vertex('5')
grafo4.add_vertex('6')
grafo4.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
grafo4.vs['name'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
plot(grafo4, bbox = (0,0,300,300))
```

```python
# Impressão da matriz de adjacência.

print(grafo4.get_adjacency())
```

```python
# Linha

print(grafo4.get_adjacency()[0,])
```

```python
# Linha e coluna

grafo4.get_adjacency()[0,1]
```

```python
# Estrutura de repetição (for) para percorrer cada vértice, visualizando o nome e o rótulo.

for i in grafo4.vs:
    print(i)
```

```python
# Criação do grafo com peso entre as relações.

grafo5 = Graph(edges = [(0,1), (2,3), (0,2), (0,3)], directed = True)
grafo5.vs['label'] = ['Fernando', 'Pedro', 'José', 'Antonio']
grafo5.vs['peso'] = [20,30,40,50]
print(grafo5)
```

```python
# Percorrer os vértices para visulizar os pesos.

for v in grafo5.vs:
    print(v)
grafo5.vs[0]
```

```python
# Definição do tipo de amizade e do peso das relações.

grafo5.es['TipodeAmizade'] = ['Amigo', 'Inimigo', 'Inimigo', 'Amigo']
grafo5.es['weight'] = [1,2,1,4]
print(grafo5)
```

```python
# Percorrer os vértices tipo de amizade.

for i in grafo5.es:
    print(i)
grafo5.es[0]
```
