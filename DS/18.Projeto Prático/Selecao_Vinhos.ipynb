```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import datasets
from scipy import stats
```

```python
df_vinhos = datasets.load_wine()
stats.describe(df_vinhos.data)
```

```python
# Classificando em classe e em dados.

df_vinhos.target
```

```python
df_vinhos.data
```

```python
# Criação dos previsores.

previsores = df_vinhos.data
classe = df_vinhos.target
```

```python
# Divisão em treino e teste.

X_teste, X_treino, Y_teste, Y_treino = train_test_split(previsores, classe, random_state=0, test_size=0.3)
```

```python
gnb = GaussianNB()
gnb.fit(X_treino, Y_treino)
```

```python
# Previsões

prever = gnb.predict(X_teste)
prever
```

```python
# Matriz de confusão.

confusao = confusion_matrix(Y_teste, prever)
confusao
```

```python
# Taxa de acerto.

acerto = accuracy_score(Y_teste, prever)
acerto
```
