```python
# Conversão dos dados strings para data.


base = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/airpassengers.csv', parse_dates= ['Month'],
                   index_col='Month', date_format='%Y-%m') # date_format é usado na transofrmação de sting para data.
base
```

```python
# Tipos de atributos

print(base.dtypes)
```

```python
# Visualização do índice do data frame.

base.index
```

```python
# Criação da série temporal.

ts = base['#Passengers']
ts
```

```python
# Tipos de atributos

print(ts.dtypes)
```

```python
# Visualização um índice específico.

ts.iloc[1]
```

```python
# Visualização por ano e mês.

ts['1949-02']
```

```python
# Visualizando intervalor.

ts['1950-01-01':'1950-07-31']
```

```python
#visualização da série temporal completa.

plt.plot(ts)
```

```python
# Fazendo o gráfico separando os anos. 

ts_anual = ts.resample('YE').sum()
plt.plot(ts_anual)
```

```python
# Visualização por mês.

ts_mes = ts.groupby([lambda x: x.month]).sum()
plt.plot(ts_mes)
```

```python
# Visualização por datas específicas.

ts_datas = ts['1960-01-01':'1960-12-01']
plt.plot(ts_datas)
```
