```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from datetime import datetime
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Carregamento da base de dados, criação do atributo data e criação de série temporal.

base = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/airpassengers.csv')
base
```

```python
base = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/airpassengers.csv', parse_dates=['Month'], index_col= 'Month',
                   date_format='%Y-%m')


ts = base['#Passengers']
```

```python
# Visualização da série temporal.

plt.plot(ts)
```

```python
# Decomposição da série temporal, criando uma variável para cada formato.

decomposicao = seasonal_decompose(ts)
```

```python
# Tendência.

tendencia = decomposicao.trend
tendencia
```

```python
# Sazonalidade.

sazonal = decomposicao.seasonal
sazonal
```

```python
# Erro

erro = decomposicao.resid
erro
```

```python
# Visualização do gráfico para cada série temporal.

plt.plot(tendencia)
```

```python
plt.plot(sazonal)
```

```python
plt.plot(erro)
```

```python
plt.subplot(4,1,1)
plt.plot(ts, label = 'Original')
plt.legend(loc = 'best')

# Visualização da tendência
plt.subplot(4,1,2)
plt.plot(ts, label = 'tendencia')
plt.legend(loc = 'best')

# Visualizando a sazonalidade.
plt.subplot(4,1,3)
plt.plot(ts, label = 'sazonal')
plt.legend(loc = 'best')

# Visualização do erro.
plt.subplot(4,1,4)
plt.plot(ts, label = 'erro')
plt.legend(loc = 'best') # matplotlib seleciona o melhor lugar para a legenda.
plt.tight_layout() # ajusta o layout.
```
