```python
# Resumo.

"""
Conjunto de produtos cujas restrições, presentes no modelo relacional, são flexibilizadas:
    -Desnormalizado.
    -Sem restrições de integridade.
    -Modelo transacional mais flexível.

Os produtos buscam resolver problemas específicos de certas aplicações que eram mais difíceis de resolver no modelo relacional.
    -Processamento eficiente.
    -Paralelismo.
    -Escalabilidade.
    -Custo.


4 Principais tipos de bancos relacionais:
    -Key-Value Database (KVD).
    -Colunas Ordenadas.
    -Banco de Dados Documentos.
    -Banco de Dados Grafos.

Key-Value Pair (KVP):
    Armazena conjunto Chave/Valor.
    Em memória, disco ou híbrido.
    (Redis/Amazon/Couchbase)

Colunas Ordenadas:
    Dados orientados a colunas, onde cada linha tem um ponteiro.
    Armazenados de maneira contínua.
    Torelante a falhas atráves de sistema de arquivos distribuídos.
    Baseado no modelo Bigtable (Google).
    (Hypertable/ApacheHbase)

Banco de Dados Grafos:
    Aplicação em redes sociais, medicina, genética, RH e etc.
    (Neo4j)

Banco de Dados de Documentos:
    Não é um sistema ECM.
    Exemplo: Suporte a JSon.
    (MongoDB/CouchDB)

"""
```

```python
# Banco de Dados Documentos.

"""
Armazenamento de chave/valor.
Diferente do Key-Value Pair, o valor é um documento estruturado e indexado, com metadados.
Valor (docuemto), pode ser consultado.

Normalmente no formato JSon.
    Estrutura nome/valor, entre aspas duplas.
    Dados separados por vírgula.
    Chaves separam objetos.
    Vetores entre colchetes.
"""
```

```python
# MongoDB

"""
Open Source.
Multiplataforma.
Escalável.
Orientado a Documento: Json.
Permite documentos aninhados.
Indexados: Podem ser realizadas buscas pelo conteúdo do documento.
Não tem Schema fixo.
Não tem integridade referencial.


Principais tipos de dados:
    String
    Inteiro
    Booleano
    Double
    Array
    Timestamp
    Object
"""
```
