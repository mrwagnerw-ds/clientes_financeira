```python
from scipy.stats import norm
```

```python
# Conjunto de objetos numa cesta, a média é 8 e o desvio é 2.
# Qual a probabilidade de tirar um objeto que o peso seja menor que 6?

norm.cdf(6,8,2)
```

```python
# Qual a probabilidade de tirar um objeto maior que 6 quilos?

norm.sf(6,8,2)
```

```python
# Qual a probabilidade de tirar um objeto maior que 6 quilos? Outra forma.

1 - norm.cdf(6,8,2)
```

```python
# Qual a probabilidade de tirar um objeto que o peso é menor que 6 e maior que 10?
norm.cdf(6,8,2) + norm.sf(10,8,2)
```

```python
# Qual a probabilidade de tirar um objeto maior que 8 e menor que 10?
norm.cdf(10,8,2) - norm.cdf(8,8,2)
```
