```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# Criação de uma variável com dados em uma distribuição normal com a função rvs.

dados = norm.rvs(size = 1000)
dados
```

```python
# Histograma.

plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
# Geração de gráficos para verificar se é uma distribuição normal.

fig, ax = plt.subplots()
stats.probplot(dados, fit = True, plot = ax)
plt.show()
```

```python
# Execução do teste de Shapiro

stats.shapiro(dados)
```

```python
# Dados não normais.

dados2 = skewnorm.rvs(4, size = 1000)
```

```python
# Histograma.

plt.hist(dados2, bins = 20)
plt.title('Dados2')
```

```python
# Geração do gráfico para teste de normalidade.

fig, ax = plt.subplots()
stats.probplot(dados2, fit = True, plot = ax)
plt.show()
```

```python
# Teste de Shapiro.

stats.shapiro(dados2)
```
