```python
from sklearn import datasets
from sklearn.metrics import confusion_matrix
import numpy as np
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.cluster import cluster_visualizer
```

```python
# Carregamento da base de dados.

iris = datasets.load_iris()
```

```python
# Configuração dos parâmetros K-medoids, utilizando apenas duas colunas do banco de dados.

cluster = kmedoids(iris.data[:, 0:2], [3, 12, 20]) # 3, 12, 20 são registros aleatórios da base de dados.

# Visualização.

cluster.get_medoids()
```

```python
# Aplicando agrupamento (cluster), obtenção de previsão e visualização dos medoides.

cluster.process()
previsoes = cluster.get_clusters()
medoides = cluster.get_medoids()

# Lista de 3 elementos com os índices dos registros do cluster
previsoes
```

```python
# Visualização do agrupamento.

v = cluster_visualizer()
v.append_clusters(previsoes, iris.data[:,0:2])
v.append_cluster(medoides, data = iris.data[:,0:2], marker='*', markersize= 20)
v.show()
```

```python
# Código para criar duas listas, uma com os valores reais e outras com os grupos reais.
# Para uso da matriz de contingência(confusão).

lista_previsoes = []
lista_reais = []

for i in range(len(previsoes)):
    for j in range (len(previsoes[i])):
        lista_previsoes.append(i)
        lista_reais.append(iris.target[previsoes[i][j]])
```

```python
# Geração da matriz de confusão, comparando grupos reais e previstos.

lista_previsoes = np.array(lista_previsoes)
lista_reais = np.asarray(lista_reais)
resultados = confusion_matrix(lista_reais, lista_previsoes)
resultados
```
