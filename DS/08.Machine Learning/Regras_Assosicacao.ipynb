```python
import pandas as pd
from apyori import apriori
```

```python
# Leitura das transações.

dados = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/transacoes.txt')
dados
```

```python
# Transformação para lista, exigido pela biblioteca apyori, 6 é a quantidade de itens na base de dados.

transacoes = []
for i in range(0,5):
    transacoes.append([str(dados.values[i,j]) for j in range(0,3)])
transacoes
```

```python
# Execução do algoritmo apriori para geração de regras de associação, definindo parâmetros de suporte e confiança.

regras = apriori(transacoes, min_support = 0.5, min_confidence = 0.5, min_length = 2)
```

```python
# Criando uma variável para armazenar somente as regras de associação.

resultados = list(regras)
print(resultados[0])
#resultados
```

```python
# Criação de uma nova variável, percorrendo a variável anterior para melhor visualização dos resultados.

resultados2 = [list(x) for x in resultados]
resultados2
```

```python
# Criando nova variável para visualização das regras, adicionando as regras da variável anterior resultados2.

resultados3 = []
for i in range(0,5):
    resultados3.append([list(x) for x in resultados2[i][2]])

resultados3
```
