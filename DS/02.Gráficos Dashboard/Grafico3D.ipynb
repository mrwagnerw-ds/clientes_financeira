```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
base = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/orchard.csv')
base.head()
```

```python
# Criando o gráfico 3d passando 3 atributos (decrease, rowpos, colpos).

figura = plt.figure(layout = 'tight')
eixo = figura.add_subplot(1,1,1, projection = '3d') # Indicando o gráfico 3d.
eixo.scatter(base.decrease, base.rowpos, base.colpos)
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')
```
