```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/trees.csv')
base.shape
```

```python
# Gerando Boxplot apenas com o atributo Volume.

plt.boxplot(base.Volume, vert = False, showfliers= False, notch = True, patch_artist= True)
plt.title('Árvores')
plt.xlabel('Volume')
```

```python
# Dados por linha

plt.boxplot(base)
plt.title('Árvores')
plt.xlabel('Dados')
```

```python
# Gerando 3 boxplots com atributos diferentes.

plt.boxplot(base.Volume, vert=False)
plt.boxplot(base.Girth, vert = False)
plt.boxplot(base.Height, vert = False)
plt.title('Árvores')
plt.xlabel('Dados')
```
