```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/co2.csv')
base.head()
```

```python
# Criando variáveis para cada atributo.

x = base.conc
y = base.uptake
```

```python
# Retorna os valores únicos do atributo "treatment".

unicos = list(set(base.Treatment))
unicos
```

```python
# Loop que percorre cada tipo de tratamento (chilled e nonchilled) e cria um gráfico de dispersão.

for i in range(len(unicos)):
    indice = base.Treatment == unicos[i]
    plt.scatter(x[indice], y[indice], label = unicos[i])
plt.legend(loc = 'lower right')
```
