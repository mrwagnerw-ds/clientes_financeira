```python
import pandas as pd
```

```python
base = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/insect.csv')
base.shape
```

```python
# Agrupando atributos contando e somando.

grupo = base.groupby(['spray'])['count'].sum()
grupo
```

```python
# Gerando gráficos de barras.
# Para criar gráficos de barra ou de pizza é necessário o agrupamento dos dados (código anterior).


grupo.plot.bar(color = 'blue')
```

```python
# Mudando as cores das barras.

grupo.plot.bar(color = ['blue', 'gray', 'red', 'magenta', 'green', 'yellow'])
```

```python
# Gráfico de pizza.

grupo.plot.pie()
```

```python
# Gráfico de pizza com legendas.

grupo.plot.pie(legend = True)
```
