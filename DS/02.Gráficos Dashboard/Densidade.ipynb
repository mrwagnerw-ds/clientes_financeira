```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
# Carregamento dos dados.

base = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/trees.csv')
base.head()
```

```python
# Histograma

plt.hist(base.iloc[:,1], bins = 6)
```

```python
# Histograma com linha de densidade (kde).

sbn.histplot(base.iloc[:,1], kde = False, bins = 6, color = 'blue').set(title = 'Árvores')
```

```python
# Gráfico de densidade e histograma.

sbn.histplot(base.iloc[:,1], kde = True, bins = 6, color = 'blue').set(title = 'Árvores')
```

```python
# Usando o método kdeplot.

sbn.kdeplot(base.iloc[:,1], color = 'red').set(title = 'Árvores')
```
