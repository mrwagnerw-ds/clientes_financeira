```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
base = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/trees.csv')
base.shape
```

```python
# Gráfico de dispersão considerando volume e a dispersão.

plt.scatter(x=base.Girth, y=base.Volume, color = 'blue', facecolors = 'none', marker = '*')
plt.title('Árvores')
plt.ylabel('Circunferência')
plt.xlabel('Volume')
```

```python
# Grafico de linha considerando o volume e o atributo 'Girth', sem o marker.

plt.plot(base.Girth, base.Volume)
plt.title('Árvores')
plt.ylabel('Circunferência')
plt.xlabel('Volume')
```

```python
# Gráfico de dispersão com afastamento de dados (jitter).
# fit_reg linha de tendência.

sbn.regplot(x = base.Girth, y = base.Volume, data = base, x_jitter = 0.3, fit_reg= False)
```

```python
# O mesmo código do exemplo anterior mas com o fit_reg True.

sbn.regplot(x = base.Girth, y = base.Volume, data = base, x_jitter = 0.3, fit_reg= True)
```
