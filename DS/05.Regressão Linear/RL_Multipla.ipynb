```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
```

```python
# Carregamento da base de dados e exclusão da primeira coluna.

base = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/mt_cars.csv')
base.head()
```

```python
# Exclusão da primeira coluna.

base = base.drop(['Unnamed: 0'], axis = 1)
base.head()
```

```python
# Criação das variáveis dependentes e independentes.

X = base.iloc[:, 2].values # coluna disp.
Y = base.iloc[:, 0].values # coluna mpg.

correlacao = np.corrcoef(X, Y)
correlacao
```

```python
# Mudança do formato de X para o formato de matriz.

X = X.reshape(-1, 1)
```

```python
# Criação do modelo e treinamento com o comando fit().

modelo = LinearRegression()
modelo.fit(X, Y)
```

```python
# Visualização de interceptação.

modelo.intercept_
```

```python
# Visualização de inclinação.

modelo.coef_
```

```python
# visualização do score.

modelo.score(X, Y)
```

```python
# Visualização das previsões.

previsoes = modelo.predict(X)
previsoes
```

```python
# Criação do modelo com a biblioteca statsmodels.

ajuste = sm.ols(formula = 'mpg ~ disp', data = base)
treinamento = ajuste.fit()
treinamento.summary()
```

```python
# Visualizando os resultados.

plt.scatter(X, Y)
plt.plot(X, previsoes, color = 'red')
```

```python
# Prevendo um único valor.

modelo.predict([[200]])
```

```python
# Criação de novas variáveis X1 e Y1 e novo modelo de para comparação com a anterior.

X1 = base.iloc[:, 1:4].values
X1
```

```python
Y1 = base.iloc[:, 0].values
modelo2 = LinearRegression()
modelo2.fit(X1, Y1)
```

```python
# Visualização do score.

modelo2.score(X1, Y1)
```

```python
# Criando ajuste com mais atributos (regressão linear múltipla).
# statsmodels.

ajuste2 = sm.ols(formula = 'mpg ~ cyl + disp + hp', data = base) # adj.r-squared mais alto significa que o modelo é melhor explicado na relação entre os atributos.
treinamento2 = ajuste2.fit()
treinamento2.summary()
```

```python
# Previsão de um novo registro.

novo = np.array([4, 200, 100])
novo = novo.reshape(1, -1)
modelo2.predict(novo)
```
