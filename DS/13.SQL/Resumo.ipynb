```python
# Structured Query Language

"""
Linguagem de consulta para dados relacionais/estruturados.
Amplamente popular.
Padronizada/Declarativa.
Especialmente eficiente.

Consultar dados:
DQL - Select

Manipulação de dados:
DML - Insert
      Update
      Delete

Definição de dados:
DDL - Create Table / Index / View
      Alter Table / Index
      Drop Index / View

"""
```

```python
# DDL - Linguagem de definição de dados.

"""
Create sequence = cria uma sequência de dados.

Create table clientes = cria uma tabela de cliente com suas informações atribuindo uma Key Primary (ID).

"""
```

```python
# Select

"""
Estrutura básica = select * from tabela

select = comando
* = colunas (todas)
from = comando
tabela = tabela de trabalho.

Filtro:

where = usado pra definir um filtro, incluindo um atributo para ser especificado.
or = condicional para agregar filtros.
in = um operador de conjunto, trazendo tudo que está dentro do conjunto.
like = usado para buscar texto de forma parcial (%alb%_) - precisa dos percentuais e do underline.
> maior que = usado para comparativos numéricos.
order by = usado para ordenar informações.
order by + desc = ordenar por ordem decrescente.
between = significa entre, usado para trazer valores dentro de um intervalo.
limit = estipula o limite de dados que será acessado.
distinct = traz valores únicos.
max() = traz o valor máximo.
min() = traz o valor mínimo.
avg() = traz a média.
sum() = traz a soma.
group by = agrupa por atributo.
having = usado junto com o group by para especificar uma condição para o atributo.



"""
```

```python
# Join

"""
Usado para unir informações.

Inner Join = retorna os registro equivalentes entre tabelas.
Left Join = os dados que existam equivalência e os dados que não tem.
Right Join = Funciona da mesma forma que o left join, mas trazendo informações da tabela da direita.
Full Join = traz registros das duas tabelas.

"""
```
