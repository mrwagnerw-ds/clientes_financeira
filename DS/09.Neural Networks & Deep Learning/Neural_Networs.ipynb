```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.utils import to_categorical
import numpy as np
 
```

```python
# Carregamento da base de dados e criaГ§ГЈo dos previsores e classes.

base = datasets.load_iris()
previsores = base.data
classe = base.target
classe
```

```python
# TransformaГ§ГЈo da classe para dummy, pois temos uma rede neural com 3 neurГҙnios.

classe_dummy = to_categorical(classe)
#classe_dummy
```

```python
# Dividindo a base de dados em treino e teste.

X_treino, X_teste, Y_treino, Y_teste = train_test_split(previsores, classe_dummy, test_size= 0.3, random_state=0)
```

```python
# CriaГ§ГЈo do modelo neural com o comando Sequential.
modelo = Sequential()

# A definiГ§ГЈo da entrada Г© feita fora das camadas.
modelo.add(Input(shape = (4,)))

# Primeira camada oculta com 5 neurГҙnios.
modelo.add(Dense(units = 5, input_dim = 4))

# Segunda camada oculta
modelo.add(Dense(units = 4))

# Terceira camada oculta / FunГ§ГЈo softmax por que temos um problema de classificaГ§ГЈo com mais de duas classes.
modelo.add(Dense(units=3, activation= 'softmax'))
```

```python
# VisualizaГ§ГЈo da estrutura da rede neural.

modelo.summary()
```

```python
# ConfiguraГ§ГЈo dos parГўmetros da rede neural (adam = atualiza os pesos / loss = calcula o erro).

modelo.compile(optimizer='adam', loss = 'categorical_crossentropy',
               metrics = ['accuracy'])

# Treinamento dividindo a base de treinamento em uma porГ§ГЈo de validaГ§ГЈo.
modelo.fit(X_treino, Y_treino, epochs = 1000, validation_data = (X_teste, Y_teste))
```

```python
previsoes = modelo.predict(X_teste)
previsoes = (previsoes > 0.5)
previsoes
```

```python
# Como Г© um problema com trГӘs saГӯdas, precisamos buscar a posiГ§ГЈo que possui o maior valor

YT_matrix = [np.argmax(t) for t in Y_teste]
YP_matrix = [np.argmax(t) for t in previsoes]
```

```python
confusao = confusion_matrix(YT_matrix, YP_matrix)
confusao
```

```python
from sklearn.metrics import accuracy_score

acerto = accuracy_score(YT_matrix, YP_matrix)
acerto
```
