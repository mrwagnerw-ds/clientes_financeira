```python
from scipy.stats import t
```

```python
# Se a média de salário dos cientistas de dados é de R$ 75 reais por hora, amostra com 9 funcionário e desvio padrão é 10.
# Qual a probabilidade de selecionarmos um cientista de dados e o salário ser menor que R$ 80 reais?

t.cdf(1.5, 8)
```

```python
# Qual a probabilidade do salário ser maior que R$ 80 reais?

t.sf(1.5, 8)
```

```python
# Somatório das funções dos valores de antes e de depois.

t.cdf(1.5, 8) + t.sf(1.5, 8)
```
