```python
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import MultiComparison
```

```python
# Carregamento da base de dados.

tratamento = pd.read_csv('C:/Users/mrwag/OneDrive/Big Data/DS/21.Textos/anova.csv', sep = ';')
tratamento
```

```python
# Boxplot agrupando os dados dos remédios.

tratamento.boxplot(by = 'Remedio', grid = False)
```

```python
# Boxplot agrupando os dados dos remédios.

tratamento.boxplot(by = 'Remedio', grid = True) # Com grade.
```

```python
# Criação do modelo de regressão linear e execução do teste.

modelo1 = ols('Horas ~ Remedio', data = tratamento).fit()
resultados1 = sm.stats.anova_lm(modelo1)

# Observar se o valor de p é maior que 0.5 para aceitar a hipótese nula.

resultados1
```

```python
# Criação do modelo com mais atributos e execução do teste.

modelo2 = ols('Horas ~ Remedio * Sexo', data = tratamento).fit()
resultados2 = sm.stats.anova_lm(modelo2)

# Observar se o valor do P é menor que 0.5 para rejeitar a hipótese nula.

resultados2
```

```python
# Havendo diferença o teste de Tukey é executado.
# Execução do teste e visualização do gráfico.

mc = MultiComparison(tratamento['Horas'], tratamento['Remedio'])
resultado = mc.tukeyhsd()
print(resultado)
resultado.plot_simultaneous()
```
