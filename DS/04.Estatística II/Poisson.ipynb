```python
from scipy.stats import poisson
```

```python
# Se a média de acidentes de carro é 2 ao dia.
# Qual a probabilidade de ocorrerem 3 ao dia?

poisson.pmf(3, 2)
```

```python
# Qual a probabilidade se ocorrerem 3 ou menos?

poisson.cdf(3, 2)
```

```python
# Qual a probabilidade de ocorrerem mais de 3 acidentes ao dia?

poisson.sf(3, 2)
```
