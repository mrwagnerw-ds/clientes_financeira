```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
# Criação da matriz e execução do teste.

novela = np.array([[19, 6], [43,32]])
novela
```

```python
# Segundo valor é o pvalue.
# O valor de p sendo maior que 0.5 não há evidências de diferença significativa.

chi2_contingency(novela)
```

```python
novela2 = np.array([[22, 3], [43,32]])
novela2
```

```python
chi2_contingency(novela2) # pvalue menor que 0.5 aceitamos a hipótese alternativa, há sim diferença significativa.
```
