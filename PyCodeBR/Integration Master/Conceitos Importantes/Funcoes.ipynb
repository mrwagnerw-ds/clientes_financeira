```python
def envia_email(nome, e_mail):
    nome_destino = nome
    email_destino = e_mail
    return f'Email Enviado para {nome_destino} de e-mail {email_destino}'
```

```python
pessoas = [
    {
    'nome': 'Wagner',
    'email': 'wagner@gmail.com'},
    {'nome': 'Juan',
    'email': 'juan@gmail.com'},
    {'nome': 'Luna',
    'email': 'luna@gmail.com'},
    {'nome': 'Malu',
    'email': 'Malu@gmail.com'},
    {'nome': 'Morgana',
    'email': 'morgana@gmail.com'
    }
]

for pessoa in pessoas:
    email_enviado = envia_email(pessoa['nome'],pessoa['email'])
    print(email_enviado)
```
