```python
"""
Usar o loop for quando tenho uma noção de quantas vezes meu código irá rodar.

Usar o loop while quando eu tenho uma condição e preciso rodar meu código até a condição ficar False.
"""

def envia_email(cliente):
    print(f'E-mail Enviado para o cliente {cliente}!')
```

```python
clientes = ['Wagner', 'Juan', 'Luna', 'Malu', 'Morgana']

for n in clientes: # Caso queira indicar o índice usar a função enumerate.
    envia_email(n)
```

```python
numero = 0

while numero <= 5:
    print(numero)
    numero += 1
```

```python
for i in range(10):
    print(i)
```

```python
for i, cliente in enumerate(clientes):
    if i == 2:
        break
    envia_email(cliente)
```

```python
"""
Break = Para toda a função na condição indicada.

Continue = Roda a função mas pula os parâmetros indicados dentro da função, passando para a próxima condição.


"""

num = 0

while num < 10:
    num += 1
    if num ==5:
        continue
    print(num)
```

```python
for i, cliente in enumerate(clientes):
    if i == 2:
        continue
    envia_email(cliente)
```
