```python
a = True
b = False
```

```python
# And - Necessita que as duas condições sejam verdadeiras.


if a and b:
    print('Atendeu a condição!')
else:
    print('Não atendeu a condição!')
```

```python
c = True
d = True

if c and d:
    print('Atendeu a condição!')
else:
    print('Não atendeu a condição!')
```

```python
# Or - Um precisa ser verdadeiro.

if a or b:
    print('Atendeu a condição!')
else:
    print('Não atendeu a condição!')
```

```python
if c and d:
    print('Atendeu a condição!')
else:
    print('Não atendeu a condição!')
```

```python
e = False
f = False

if e and f:
    print('Atendeu a condição!')
else:
    print('Não atendeu a condição!')
```

```python
# == - Compara em igualdade.

nome = 'Felipe'
idade = 24

if nome == 'Felipe' and idade == 25:
    print('Dados corretos!')
else:
    print('Dados incorretos!')
```

```python
# != - Compara as diferenças.

if nome == 'Felipe' or idade != 25:
    print('Dados corretos!')
else:
    print('Dados incorretos!')
```

```python
nome = 'Wagner'
idade = 34
peso = 90


if (nome == 'Wagner' or idade == 25) and peso == 90:
    print('Dados corretos!')
else:
    print('Dados incorretos!')
```
