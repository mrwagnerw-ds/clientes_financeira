```python
numeros = [1, 2, 3, 4, 5]
```

```python
numeros_dobrados = []

for i in numeros:
    numeros_dobrados.append(i * 2)
```

```python
numeros = [1, 2, 3, 4, 5]

def dobro(numero):
    return numero * 2

numeros_2vezes = [dobro(numero) for numero in numeros] # [expressão ou função + loop for]
```

```python
numeros_2vezes
```

```python
nomes = ['Ana', 'Felipe', 'Joao', 'Arlindo', 'Carlos']

nomes_maiusculos = [nome.upper() for nome in nomes]

nomes_maiusculos
```

```python
nomes = ['Ana', 'Felipe', 'Joao', 'Arlindo', 'Carlos']

nomes_maiusculos = [nome.upper() for nome in nomes if nome[0] =='A'] # Epressão/função + loop for + condicionais.

nomes_maiusculos
```
