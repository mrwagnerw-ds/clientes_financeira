```python
texto = 'AULA PYCODEBR'
```

```python
# Sintaxe [] - busca o caractere pelo índice dentro do colchetes.

texto[3]
```

```python
# COUNT - Conta a recorrência de determinado caractere.

texto.count('A')
```

```python
# Passando o índice da posição de início e fim da recorrência.

texto.count('P', 5, 11)
```

```python
# Len - conta a quantidade de caracteres.

len(texto)
```

```python
# Upper() - função para colocar as letras em caixa alta.

texto.upper()
```

```python
# Lower() - função para colocar as letras mínusculas.

texto.lower()
```

```python
# Capitalize() - Coloca apenas a primeira letra do texto em maiúscula.

texto.capitalize()
```

```python
# Title() - Coloca a primeira letra de cada palavra em maíuscula.

texto.title()
```

```python
# Find() - Traz o índice onde inicia a busca.

texto.find('AULA')
```

```python
# Split() - separa as palavras do texto em lista.

texto.split()
```

```python
# Join() - Agrupa as palavras usando o separador dentro das aspas, caso não tenha separador, ele apenas junta tudo.

palavras = texto.split()
' '.join(palavras)
```

```python
# Strip() - Elemina os espaços em branco antes e após o texto.
# Rstrip() - Elimina os espaços em branco a direito do texto.
# Lstrip() - Elimina os espaços em branco a esquerda do texto.

texto = '      AULA PYCODEBR     '
texto.strip()
```
