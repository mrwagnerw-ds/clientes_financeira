```python
"""
Método de requisição https:

O protocolo HTTP define um conjunto de métodos de requisição responsáveis por indicar a ação a ser executada para um dado recurso. Embora esses métodos
possam ser descritos como substantivos, eles também são comumente referenciados como HTTP Verbs. Cada um deles implementa uma semântica diferente, mas alguns
recursos são compartilhados por um grupo deles, como por exemplo, qualquer método de requisição pode ser do tipo safe, idempotent ou cacheable.


Get:
Solicita a representação de um recurso específico, devendo retornar apenas dados. Método seguro, pois é só um método de leitura.

Head:
Solicita uma resposta idêntica ao GET mas sem o corpo da resposta.

Post:
É utilizado para submeter uma entidade a um recurso específico, enviando dados. Corre o risco de alterar dados no servidor.

Put:
Quando eu busco alterar os dados do meu objeto já cadastrado. Usado para subescrever todos os dados.

Patch:
Usado para alterações parciais dos recursos. Usado para alguns campos dos dados.

Delete:
Usado para deletar campos dos meus dados.

Numeros do Status:
100 -> 199 : Dados trazido pelo status são meramente informativos.
200 -> 299 : Dados de sucesso na resposta.
300 -> 399 : Redirecionamento de mensagem.
400 -> 499 : Erros referente ao uso do usuário do api.
500 -> 599 : Erros referente ao servidor.


developer.mozilla.org/en-US/docs/Web/HTTP/Status*information_responses 

"""
```
